```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
iris.shape
```

```python
iris.info()
```

```python
iris['species'].value_counts()
```

### Line Plot (Matplotlib)

```python
plt.plot(iris['sepal_length'])
plt.xlabel("Index")
plt.ylabel("Sepal Length")
plt.title("Line Plot of Sepal Length")
plt.show()
```

### Line Plot (Matplotlib)

```python
plt.scatter(iris['sepal_length'], iris['petal_length'])
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.title("Scatter Plot")
plt.show()
```

### Scatter Plot (Matplotlib)

```python
plt.scatter(iris['sepal_length'], iris['petal_length'])
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.title("Scatter Plot")
plt.show()
```

### Scatter Plot with Hue (Seaborn)

```python
sns.scatterplot(data=iris, x="sepal_length", y="petal_length", hue="species")
plt.title("Sepal Length vs Petal Length")
plt.show()
```

### Histogram (Seaborn)

```python
sns.histplot(iris['petal_length'], bins=10, kde=True)
plt.title("Histogram of Petal Length")
plt.show()
```

### Box plot

```python
sns.boxplot(data=iris, x="species", y="sepal_length")
plt.title("Box Plot of Sepal Length by Species")
plt.show()
```

### Violin Plot

```python
sns.violinplot(data=iris, x="species", y="petal_length")
plt.title("Violin Plot of Petal Length")
plt.show()
```

### Strip Plot

```python
sns.stripplot(data=iris, x="species", y="petal_length", jitter=True)
plt.title("Strip Plot of Petal Length")
plt.show()
```

### Count Plot

```python
sns.countplot(data=iris, x="species")
plt.title("Count of Each Species")
plt.show()
```

### ECDF Plot (Empirical Cumulative Distribution)

```python
sns.ecdfplot(data=iris, x="petal_length", hue="species")
plt.title("ECDF Plot of Petal Length")
plt.show()
```

### Point Plot

```python
sns.pointplot(data=iris, x="species", y="sepal_length")
plt.title("Point Plot of Sepal Length")
plt.show()
```
