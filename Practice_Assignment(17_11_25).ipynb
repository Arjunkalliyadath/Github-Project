1. Write a Python program that takes an integer as input and calculates the
sum of its digits

```python
num = int(input("Enter a number: "))
sum_of_digits = 0

while num > 0:
  digit = num % 10
  sum_of_digits += digit
  num = num // 10
print("Sum of digits:", sum_of_digits)
```

2. Create a Python program to calculate the factorial of a given number using
a function.

```python
def factorial(n):
    fact = 1
    for i in range(1, n + 1):
        fact = fact * i
    return fact

num = int(input("Enter a number: "))
result = factorial(num)
print("Factorial of", num, "is", result)
```

3. Develop a Python program that checks whether a given word or phrase is a
palindrome.

```python
text = input("Enter a word or phrase: ")

clean_text = text.replace(" ", "").lower()

reverse_text = clean_text[::-1]

if clean_text == reverse_text:
    print("It is a palindrome.")
else:
    print("It is not a palindrome.")
```

4. Create a Python program that takes a sentence as input and counts the
number of vowels and consonants.

```python
sentence = input("Enter a sentence: ")

sentence = sentence.lower()
vowels = "aeiou"

vowel_count = 0
consonant_count = 0

for ch in sentence:
    if ch.isalpha():
        if ch in vowels:
            vowel_count += 1
        else:
            consonant_count += 1

print("Vowels:", vowel_count)
print("Consonants:", consonant_count)
```

6. Write a Python program that takes a string as input and counts the
frequency of each character in the string.


```python
text = input("Enter a string: ")

freq = {}

for ch in text:
    if ch in freq:
        freq[ch] += 1
    else:
        freq[ch] = 1

print("Character Frequency:")
for key, value in freq.items():
    print(key, ":", value)
```

10. Write a python program to find the roots of a quadratic equation


```python
a = float(input("Enter value of a: "))
b = float(input("Enter value of b: "))
c = float(input("Enter value of c: "))

d = (b*b) - (4*a*c)

if d > 0:
    root1 = (-b + (d ** 0.5)) / (2*a)
    root2 = (-b - (d ** 0.5)) / (2*a)
    print("Two real roots:", root1, "and", root2)

elif d == 0:
    root = -b / (2*a)
    print("One real root:", root)

else:
    print("No real roots (imaginary roots)")
```

4. Write a Python function that takes a positive integer and returns the sum
of the cube of all the positive integers smaller than the specified number.\
Sample input: 4\
Sample output: 36.

```python
def sum_of_cubes(n):
  total = 0
  for i in range(1, n):
    total = total + (i*i*i)
  return total

num = int(input("Enter a positive integer: "))
print(sum_of_cubes(num))
```

6. Write a Python program to find the most frequent item in a list of
numbers.\
Sample input: 2, 3, 4, 2, 5, 2\
Sample output: 2


```python
numbers = [2, 3, 4, 2, 5, 2]

most = numbers[0]
count_most = numbers.count(most)

for n in numbers:
  if numbers.count(n) > count_most:
    most = n
    count_most = numbers.count(n)

print(most)
```
