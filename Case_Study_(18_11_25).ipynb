1. Write function that returns mean value of a list of numbers

```python
def mean(numbers):
  sum = 0
  for n in numbers:
    sum = sum + n
  return sum/len(numbers)

Newlist = [8,6,7,9]
print("Mean:", mean(Newlist))
```

3. Write a function that returns the median value of a list of numbers

```python
def median(List_numbers):
  total_elements = len(List_numbers)
  sorted_numbers = sorted(List_numbers)
  mid = total_elements // 2

  if total_elements % 2 == 1:
    return sorted_numbers[mid]
  else:
    return (sorted_numbers[mid - 1] + sorted_numbers[mid]) / 2

NewList = [2,9,4,7,3,3]
print ("Median:", median(NewList))
```

5. Write a function to return standard deviation of a list of numbers

```python
def standard_deviation(numbers):
    n = len(numbers)
    mean = sum(numbers)/n

    squared_diff = [(x - mean) ** 2 for x in numbers]
    variance = sum(squared_diff) / n

    return variance ** 0.5
x = [2,5,7,4,5,9,8,1]
print(standard_deviation(x))
```
